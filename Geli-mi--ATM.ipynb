```python
import time
l={"ad":"","şifre":""}
h=3
while True:
    print("Merhaba\ngirş yapmak için 1'i\nüye olmak için 2'yi\nTUŞLAYINIZ\n")
    x=input("tuş: ")

    if x == "q":
        print("çıkış yapıldı\n😊")
        break
    elif x=="1":
        a=input("kullanıcı adınız:  ")
        s=input("şifreniz: ")
        if(h==0):
            break
        if a==l["ad"] and s==l["şifre"]:
            print("\nhoşgeldiniz")
            b=1000
            while True:    
                print("\nişlem seçin\n1=bakiye sor\n2=para yatır\n3=para çek\nçıkmak için x tuşlayınız!\n")
                i = input("İşlemi giriniz: ")
                if (i == "x"):
                    print("güle güle")
                    break
                elif (i == "1"):
                    print("Bakiyeniz: ",b)
                elif (i == "2"):
                    y = int(input("miktar girin: "))
                    b+=y
                elif (i == "3"):
                    y = int(input("tutar: "))
                    if (b-y < 0 ):
                        print("baye yetersiz")
                        print("Bakiyeniz: ",b)
                        continue
                    b-=y
            
        else:
            print("şireniz veya kullanıcı adınız yanlış\n🙁")
            h-=1
            print("hakınız: ",h)
        
    elif x=="2":
        a=input("kullanıcı adınız: ")
        s=input("şifreniz: ")
        l["ad"]+=a
        l["şifre"]+=s
        print("sisteme kaydediliyor...\n")
        time.sleep(2)
        print("sisteme kaydedildiniz...\n")
```
